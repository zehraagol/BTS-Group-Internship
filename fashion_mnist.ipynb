```python
import tensorflow as tf
from tensorflow import keras
from PIL import Image
```

```python
femnist = keras.datasets.fashion_mnist
```

```python
(x_train, y_train),(x_test, y_test) = femnist.load_data()

x_train = keras.utils.normalize(x_train, axis=1)
x_test = keras.utils.normalize(x_test, axis=1)
```

```python
x_test.shape
```

```python
import numpy as np

np.unique(y_train)
```

```python
import matplotlib.pyplot as plt

plt.imshow(x_train[5], cmap='gray')
print(y_train[5])
```

```python
model = keras.models.Sequential()
```

```python
model.add(keras.layers.Flatten())
```

```python
model.add(keras.layers.Dense(256,activation= tf.nn.relu))
model.add(keras.layers.Dense(128,activation= tf.nn.relu))
model.add(keras.layers.Dense(10,activation= tf.nn.softmax)) #softmax varsa class sayisini veriyon ve en sonda softmaxi veriyon
```

```python
model.compile(optimizer='adam', 
              loss= 'sparse_categorical_crossentropy',
              metrics= ['accuracy'])
```

```python
#model.build(input_shape=x_train[0].shape)
```

```python
#model.summary()
```

```python
history = model.fit(x_train,y_train, validation_data=(x_test,y_test),epochs=10)
```

```python
predictions=model.predict(x_test)
```

```python
model.evaluate(x_test,y_test)
```
